```python
#Task-1

 
import random 
import math

def MIN_MAX(alpha,beta,depth,pos,miximizing,leaf): #recursive way ==miximizing: 
    #A boolean value that indicates whether the current player is maximizing (True) or minimizing (False).
    
    if depth==0 :
        return(leaf[pos]) 
  
    if miximizing==False: #the function represents the minimizing player's turn, and it 
        #iterates over the possible moves (children) from the current position. It calls the 
        #MIN_MAX function recursively with miximizing set to True for each child, and updates the min_val and beta accordingly.
        min_val=math.inf
        for i in range(2):
            val=MIN_MAX(alpha,beta,depth-1,pos*2+i,True,leaf)
            min_val=min(min_val,val)
            beta=min(beta,val)
            if beta<=alpha:
                break
        return(min_val)

    elif miximizing==True: #the function represents the maximizing player's turn, and 
        #it follows a similar process as the minimizing player but aims to maximize the max_val and alpha instead.
        max_val=-math.inf
        for i in range(2):
            val=MIN_MAX(alpha,beta,depth-1,pos*2+i,False,leaf)
            max_val=max(max_val,val)
            alpha=max(alpha,val)
            if beta<=alpha:
                break
        return(max_val)
    
######################################################


st_id=input("Enter your student ID : ") #19201027 =19281827
st_id=st_id.replace("0","8") #0 placed by 8
w=int(st_id[-1:-3:-1]) #slicing ==starting index,ending index(excluding),step  id reverse  hoye then 
total_point=w #72

m=int(total_point*1.5) #72*1.5
max_achive=m
n=int(st_id[4]) #1
min_achive=n

suf_time=int(st_id[3]) #8 leaf node
lst=[]
for i in range(suf_time):
    val=random.randint(min_achive,max_achive) #1-108
    lst.append(val)

point_achive=MIN_MAX(-math.inf,math.inf,3, 0,True,lst) #alpha beta purning
print()
print("Task1 output")
print()
print("Generated 8 random points between the minimum and maximum point limits:",lst) 
print("Total points to win:",total_point) 
print("Achieved point by applying alpha-beta pruning =",point_achive)
if point_achive>=total_point:
    print("The winner is Optimus Prime")
else:
    print("The Winner is Megatron ")


#############################

#Task-2

shuffle_lst=[] #list er sequence changing store here
print()
print("Task2 output")
print("After the shuffle:")

for i in range(suf_time):
    random.shuffle(lst)
    point_achive=MIN_MAX(-math.inf,math.inf,3, 0,True,lst)
    shuffle_lst.append(point_achive)

print("List of all points values from each shuffle:",shuffle_lst)
print("The maximum value of all shuffles:",max(shuffle_lst))
count=0
for i in shuffle_lst:
    if i>=total_point:
        count=count+1
print("Won "+str(count)+" times out of "+str(suf_time)+" number of shuffles ")
```
